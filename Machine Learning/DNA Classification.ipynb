A promoter is a sequence of DNA needed to turn a gene on or off. The process of transcription is initiated at the promoter. Usually found the beginning of a gene, the promoter has a binding site for the enzyme used to make a messenger RNA (mRNA) molecule.<br>
Source: Google

```python
import pandas as pd
data=pd.read_csv(r"https://archive.ics.uci.edu/ml/machine-learning-databases/molecular-biology/promoter-gene-sequences/promoters.data")
```

```python
data.columns=["class","id","Sequence"]
```

```python
data
```

According to this dataset:

<li>"+" means the promoters</li>
<li>"-" means is anything that is not promoters.</li>
DNA molecules as a linear, non-overlapping sequence of the nitrogenous bases Adenine (A), Guanine (G), Cytosine (C) and Thymine (T). These are the "alphabet" of letters that are used to write the "code words".

```python
d1=data.copy()
```

```python
y=d1.pop("class")
```

```python
d1
```

```python
x={}
start=1
for i in d1["Sequence"]:
    w1=i.replace("\t","")
    w2=list(w1)
    x[start]=w2
    start+=1
```

```python
len(x[1])
```

```python
x.keys()
```

```python
import numpy as np
loc=pd.DataFrame(x)
DNA_codes=loc.T
```

```python
DNA_codes.shape
```

```python
DNA_codes.shape
```

```python
pd.get_dummies(DNA_codes) # sparse matrix of DNA_codes dataset.
```

```python
k=x.keys()
new_data={}
for i in k:
    "Enconding string indicators to numeric"
    "a:0"
    "c:1"
    "g:3"
    "t:4"
    new_data[i]=[]
    for j in x[i]:
        if j=="a":
            new_data[i].append(0)
        elif j=="c":
            new_data[i].append(1)
        elif j=="g":
            new_data[i].append(2)
        elif j=="t":
            new_data[i].append(3)
#print(new_data)
data_new=pd.DataFrame(new_data).T
data_new.head()   
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
y_encode=le.fit_transform(y)
```

```python
y_encode
```

```python
y_encode.shape
```

```python
data_new["target"]=y_encode
```

<h3>BOOsting Algorithms 1: Adaaptive boosting </h1>

AdaBoost classifier builds a strong classifier by combining multiple poorly performing classifiers 
so that you will get high accuracy strong classifier.<b> (Week learner convert into strong learner)</b> <br>
<b>Bagging </b>merging the same type of predictions. <br>
<b>Boosting</b>  merging different types of predictions(improve weak learners).<br> Bagging decreases variance, not bias, and solves over-fitting issues in a model. Boosting decreases bias, not variance.



```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model=AdaBoostClassifier(n_estimators=10) # default estimator is "DecisionTreeClassifier"
```

```python
# base estimator as DecisionTreeClassifier
model=AdaBoostClassifier(n_estimators=10)
model.fit(xtrain,ytrain)
pred=model.predict(xtest)
from sklearn.metrics import accuracy_score
acc=accuracy_score(pred,ytest)
print(acc)
```

```python
result["AdaboostClassifier"]=acc
```

```python
accf=[]
for i in range(500):
    model=AdaBoostClassifier(n_estimators=i+1)
    model.fit(xtrain,ytrain)
    pred=model.predict(xtest)
    from sklearn.metrics import accuracy_score
    acc=accuracy_score(pred,ytest)
    accf.append(acc)
plt.plot(accf)
```
